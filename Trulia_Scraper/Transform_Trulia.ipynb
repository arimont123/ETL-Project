```python
import pandas as pd
```

```python
#Load in csv
df = pd.read_csv('truliaOuput.csv')

#Get rid of unwanted columns
del df['Unnamed: 0']
df
```

```python
#Rename columns to better match description
df.columns = ['Price','City','Address','Beds','Baths','SqFt']

#Convert city column to list to modify it
city_state = df['City'].to_list()

#Split up city column into city and state
state = []
city = []
for item in city_state:
    fixed_col = item.split(',')
    state.append(fixed_col[-1].strip())
    city.append(fixed_col[-2].strip())
    try:
        district = fixed_col[-3]
    except IndexError:
        district = 'N/A'

#Get rid of strings in number columns
beds_col = []
bed = df['Beds'].to_list()
for beds in bed:
    num_beds = beds.split('b')
    beds_col.append(num_beds[0])

baths_col = []
bath = df['Baths'].to_list()
for baths in bath:
    num_bath = baths.split('b')
    baths_col.append(num_bath[0])

sqft_col = []
sqft = df['SqFt'].to_list()
for squareft in sqft:
    num_sqft = squareft.split(' ')
    sqft_col.append(num_sqft[0])
```

```python
#Add in new columns
df['City'] = city
df['State'] = state
df['Beds'] = beds_col
df['Baths'] = baths_col
df['SqFt'] = sqft_col
#Re-arrange columns
df = df[['Address','City','State','Price','SqFt','Beds','Baths',]]

#Get rid of any NA's
df.dropna(how='any')
df
```

```python
#Export cleaned csv 
df.to_csv('Cleaned_Trulia.csv')
```
