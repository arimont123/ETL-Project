```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup as BS 
import urllib.request
import ssl
import re
import pandas as pd
import numpy as np
import time 


#Cookies for the website 
def get_headers():
    
    headers = {'accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
               'accept-language':'en-US,en;q=0.9',
               'cache-control':'max-age=0',
               'upgrade-insecure-requests':'1',
               'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'}
    
    return headers

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE 

trulia_dict = {"price" : [],"city" : [],"address" : [],"bedrooms" : [],"bathrooms" : [],"houseSize" : []}


urls = ['https://www.trulia.com/TX/Austin/'] 

for url in urls: 
    
    count = 1
    
    while(count <= 100):  
        
        urlTrulia = f"{url}{count}_p/"

        print(urlTrulia)
        
        request = Request(urlTrulia, headers = get_headers())  
        htmlfile = urlopen(request) 
        htmltext = htmlfile.read() 
        soup = BS(htmltext,'html.parser') 
        
        results = soup.find_all('div',attrs = {'data-testid':'home-card-sale'}) 

        for result in results:

            price = None
            city = None
            address = None
            bedrooms = None
            bathrooms = None
            houseSize = None
            
            try:
                price = result.find('div',attrs = {'data-testid':'property-price'}).text
            except:
                pass

            try:
                city = result.find('div',attrs = {'data-testid':'property-region'}).text
            except:
                pass

            try:
                address = result.find('div',attrs = {'data-testid':'property-street'}).text
            except:
                pass

            try:
                bedrooms = result.find('div',attrs = {'data-testid':'property-beds'}).text
            except:
                pass

            try:
                bathrooms = result.find('div',attrs = {'data-testid':'property-baths'}).text
            except:
                pass

            try:
                houseSize = result.find('div',attrs = {'data-testid':'property-floorSpace'}).text
            except:
                pass

            if (price and city and address and bedrooms and bathrooms and houseSize):
                listinInfo = [price , city , address , bedrooms , bathrooms , houseSize]
                
                for key,value in zip(trulia_dict.keys(),listinInfo):
                    trulia_dict[key].append(value)

        print(count)
        print(len(trulia_dict['price']))
        print(len(trulia_dict['city']))
        print(len(trulia_dict['address']))
        print(len(trulia_dict['bedrooms']))
        print(len(trulia_dict['houseSize']))

        count += 1
        time.sleep(np.random.uniform(10, 60))
        
    
```
