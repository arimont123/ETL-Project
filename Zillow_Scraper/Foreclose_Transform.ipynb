```python
import pandas as pd
```

```python
#Load in csv
df = pd.read_csv('zillowOuput.csv')
df.head()
```

```python
#Get rid of unwanted columns
del df['Unnamed: 0']
df.head()
```

```python
#Rename columns to better match description
df.columns = ['Price','Address','Beds','Baths','SqFt','Link']

#Convert address to list to modify it
city_state = df['Address'].to_list()

#Split up city column into city and state
address = []
city = []
state_zip = []
for item in city_state:
    fixed_col = item.split(',')
    state_zip.append(fixed_col[2].strip())
    city.append(fixed_col[1].strip())
    address.append(fixed_col[0].strip())

#Split up state and zip
state = []
zip_code = []
for ad in state_zip:
    fixed_col = ad.split(' ')
    state.append(fixed_col[0].strip())
    zip_code.append(fixed_col[1].strip())

#Get rid of strings in number columns
beds_col = []
bed = df['Beds'].to_list()
for beds in bed:
    num_beds = str(beds).split('b')
    beds_col.append(num_beds[0])

#Get rid of ba in columns
baths_col = []
bath = df['Baths'].to_list()
for baths in bath:
    num_bath = str(baths).split('b')
    baths_col.append(num_bath[0])

#Get rid of est in price column
price_col = []
price = df['Price'].to_list()
for prices in price:
    if str(prices[0]) == '$':
        price_col.append(prices)
    else:
        num_price = str(prices).split(' ')
        price_col.append(num_price[1])
```

```python
#Add in new columns
df['Address'] = address
df['City'] = city
df['State'] = state
df['Zip Code'] = zip_code
df['Beds'] = beds_col
df['Baths'] = baths_col
df['Price'] = price_col

#Re-arrange columns
df = df[['Address','City','State','Zip Code','Price','SqFt','Beds','Baths','Link']]
df
```

```python
#Export cleaned csv 
df.to_csv('Foreclosed_Homes.csv')
```
